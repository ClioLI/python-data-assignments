```python
from selenium import webdriver
import time
from selenium.webdriver.chrome.options import Options
chrome_options = Options()
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument("--disable-setuid-sandbox")
browser = webdriver.Chrome('/Users/limengyu/Desktop/Raw Data #MeToo/chromedriver')
browser.set_page_load_timeout(30)   
loginurl = 'https://passport.weibo.cn/signin/login'
browser.get(loginurl)
```

```python
try:
    browser.find_element_by_id('loginName').send_keys('User ID')
    print('user success!')
except:
    print('user error!')
time.sleep(1)
```

```python
try:
    browser.find_element_by_id('loginPassword').send_keys('Password')
    print('pw success!')
except:
    print('pw error!')
time.sleep(1)
```

```python
try:
    browser.find_element_by_id('loginAction').click()
    print('click success!')
except:
    print('click error!')
time.sleep(10)
```

```python
browser.find_element_by_xpath('//div[@class="m-text-cut"]').click()
```

```python
from selenium.webdriver.common.keys import Keys
browser.find_element_by_xpath("//input[@type='search']").send_keys(u'米兔运动')
browser.find_element_by_xpath("//input[@type='search']").send_keys(Keys.ENTER) 
```

```python
names = []
times = []
contents = []
footers = []

def execute_times(times):
    for i in range(times + 1):
        browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(5)
```

```python
execute_times(100)
```

```python
for n in browser.find_elements_by_xpath('//h3[@class="m-text-cut"]'):
    name = n.text.strip()
    names.append(name)
    
for t in browser.find_elements_by_xpath('//span[@class="time"]'):
    time = t.text.strip()
    times.append(time)
    
for c in browser.find_elements_by_xpath('//div[@class="weibo-text"]'):
    content = c.text.strip()
    contents.append(content)
```

```python
for f in browser.find_elements_by_css_selector('.m-diy-btn.m-box-col.m-box-center.m-box-center-a h4'):
    footer = f.text.strip()
    footers.append(footer)

print(footers)
```

```python
#用0取代文字
new_footers = []
for i in footers:
    if '热门' in footers:
     footers.remove('热门')
    if '关注' in footers:
     footers.remove('关注')
    if i == "转发":
        i = '0'
    if i == "评论":
        i = '0'
    if i == "赞":
        i = '0'
    new_footers.append(i)


#每三个一组
step = 3
new_footers = [new_footers[i:i+step] for i in range(0,len(footers),step)]

print(new_footers)

reposts = []
for i in new_footers:
    reposts.append(i[0])


comments = []
for i in new_footers:
    comments.append(i[1])


likes = []
for i in new_footers:
    likes.append(i[2])
```

```python
import pandas as pd
import csv
with open('Weibo米兔.csv','w',newline='',encoding='utf8') as f:
    writer = csv.writer(f)
    header = ['Name','Time','Content','Repost','Comment','Like']
    writer.writerow(header)   
    writer.writerows(zip(names,times,contents,reposts,comments,likes))
```
