```python
# Read all the .txt files in folder assignment0/.
all_txt_file = ['trade-wars-news1.txt',
                'trade-wars-news2.txt',
                'trade-wars-news3.txt',
                'trade-wars-news4.txt',
                'trade-wars-news5.txt']

read_all_txt_file = []
for file in all_txt_file:
    f=open (file,'r')     
    content= f.read()
    read_all_txt_file.append(content)
    f.close()
```

```python
#For every file, remove all punctuations and extract the English words from the file content.
all_words = "".join(read_all_txt_file)

import re
punctuation='[\s+\.\!\/_,$%^*(+\"\')]+|[+——()?:【】“”‘’！，。？、~@#￥%……&*（）]+'
all_words = re.sub(punctuation,' ',all_words)

extract_words = all_words.lower().split() 
```

```python
#Handle the stop words
#Source of the stop words list: https://gist.github.com/sebleier/554280#gistcomment-8440
stop_words = ["i", "me", "my", "myself", "we", "our", "ours", "ourselves", "you", "your", "yours", "yourself", "yourselves", "he", "him", "his", "himself", "she", "her", "hers", "herself", "it", "its", "itself", "they", "them", "their", "theirs", "themselves", "what", "which", "who", "whom", "this", "that", "these", "those", "am", "is", "are", "was", "were", "be", "been", "being", "have", "has", "had", "having", "do", "does", "did", "doing", "a", "an", "the", "and", "but", "if", "or", "because", "as", "until", "while", "of", "at", "by", "for", "with", "about", "against", "between", "into", "through", "during", "before", "after", "above", "below", "to", "from", "up", "down", "in", "out", "on", "off", "over", "under", "again", "further", "then", "once", "here", "there", "when", "where", "why", "how", "all", "any", "both", "each", "few", "more", "most", "other", "some", "such", "no", "nor", "not", "only", "own", "same", "so", "than", "too", "very", "s", "t", "can", "will", "just", "don", "should", "now"]
def remove_stop_words(extract_words, stop_words):
    new_list = []
    for w in extract_words:
        if w not in stop_words:
            new_list.append(w)
    return new_list
```

```python
#For every word extracted above, count how many times each one appears
count_word={}
for word in remove_stop_words(extract_words, stop_words):
    if word not in count_word:
        count_word[word] = 1
    else:
        count_word[word] += 1
for key in count_word.keys():
  print ("%s %s " %(key , count_word[key]))
```

```python
#Rank the keywords from higher frequency to lower frequency
import operator
rank_word = sorted(count_word.items(), key=operator.itemgetter(1),reverse=True)
rank_word[0:15]
```

```python
#Output the full keyword frequency list into a CSV file. 
import csv
with open('keyword_frequency.csv','w',newline = '')as f:
    mywriter = csv.writer(f)
    header = ['word','frequency']
    mywriter.writerow(header)
    mywriter.writerows(rank_word)
```
